```python
dbname = 'imdb'

# Connect to imdb and perform some tasks
try:
    connection = psycopg2.connect(dbname=dbname,
                            user=user,
                            password=password,
                            host=host,
                            port=port)
    cursor = connection.cursor()
    
    
    # Create title_basics table
    cursor.execute("""
    CREATE TABLE midterm.title_basics(
    title_id TEXT NOT NULL,
    title_type TEXT,
    primary_title TEXT,
    original_title TEXT,
    is_adult TEXT,
    start_year TEXT,
    end_year TEXT,
    duration TEXT,
    genre TEXT);
    """)
    print(cursor.statusmessage, '\n')

    # Copy title_basics
    basics = open('title.basics.tsv', 'r') #file-like object
    copy_basics = """
           COPY midterm.title_basics(title_id, title_type, primary_title, original_title, is_adult, start_year, end_year, duration, genre) 
           FROM '/tmp/title.basics.tsv';

           """
    cursor.copy_expert(copy_basics, file=basics)
    
    
    # Create akas table
    cursor.execute("""
    CREATE TABLE midterm.akas(
    title_id TEXT NOT NULL,
    ordering TEXT,
    title TEXT,
    region TEXT,
    language TEXT,
    types TEXT,
    attributes TEXT,
    is_original_title TEXT);
    """)
    print(cursor.statusmessage, '\n')

    # Copy akas
    akas = open('title.akas.tsv', 'r') #file-like object
    copy_akas = """
           COPY midterm.akas(title_id, ordering, title, region, language, types, attributes, is_original_title) 
           FROM '/tmp/title.akas.tsv';

           """
    cursor.copy_expert(copy_akas, file=akas)
    
    
    # Create principals table
    cursor.execute("""
    CREATE TABLE midterm.principals(
    title_id TEXT NOT NULL,
    ordering TEXT,
    name_id TEXT,
    category TEXT,
    job TEXT,
    characters TEXT);
    """)
    print(cursor.statusmessage, '\n')

    # Copy principals
    principals = open('title.principals2.tsv', 'r') #file-like object
    copy_principals = """
           COPY midterm.principals(title_id, ordering, name_id, category, job, characters) 
           FROM '/tmp/title.principals.tsv';

           """
    cursor.copy_expert(copy_principals, file=principals)
    
    
    # Create crew table
    cursor.execute("""
    CREATE TABLE midterm.crew(
    title_id TEXT NOT NULL,
    directors TEXT,
    writers TEXT);
    """)
    print(cursor.statusmessage, '\n')

    # Copy crew
    crew = open('title.crew.tsv', 'r') #file-like object
    copy_crew = """
           COPY midterm.crew(title_id, directors, writers) 
           FROM '/tmp/title.crew.tsv';

           """
    cursor.copy_expert(copy_crew, file=crew)
    

    # Create episode table
    cursor.execute("""
    CREATE TABLE midterm.episode(
    title_id TEXT NOT NULL,
    parent_title_id TEXT,
    season_number TEXT,
    episode_number TEXT);
    """)
    print(cursor.statusmessage, '\n')

    # Copy episode
    episode = open('title.episode.tsv', 'r') #file-like object
    copy_episode = """
           COPY midterm.episode(title_id, parent_title_id, season_number, episode_number) 
           FROM '/tmp/title.episode.tsv';

           """
    cursor.copy_expert(copy_episode, file=episode)
    

    # Create ratings table
    cursor.execute("""
    CREATE TABLE midterm.ratings(
    title_id TEXT NOT NULL,
    average_rating TEXT,
    number_of_votes TEXT);
    """)
    print(cursor.statusmessage, '\n')

    # Copy ratings
    ratings = open('title.ratings.tsv', 'r') #file-like object
    copy_ratings = """
           COPY midterm.ratings(title_id, average_rating, number_of_votes) 
           FROM '/tmp/title.ratings.tsv';

           """
    cursor.copy_expert(copy_ratings, file=ratings)
    
    
    # Create name_basics table
    cursor.execute("""
    CREATE TABLE midterm.name_basics(
    name_id TEXT NOT NULL,
    primary_name TEXT,
    birth_year TEXT,
    death_year TEXT,
    primary_profession TEXT,
    known_for_titles TEXT);
    """)
    print(cursor.statusmessage, '\n')

    # Copy name_basics
    name_basics = open('name.basics.tsv', 'r') #file-like object
    copy_name_basics = """
           COPY midterm.name_basics(name_id, primary_name, birth_year, death_year, primary_profession, known_for_titles) 
           FROM '/tmp/name.basics.tsv';

           """
    cursor.copy_expert(copy_name_basics, file=name_basics)
       
    
    # Commit changes
    connection.commit()


except psycopg2.Error as error:
    print("Warning!", error)

finally:
        if connection is not None:
            cursor.close()
            connection.close()
```
